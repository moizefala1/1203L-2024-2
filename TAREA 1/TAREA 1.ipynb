```python

P = sp.Matrix ([[0.8, 0.2, 0.05],
                [0.05, 0.75, 0.05],
                [0.15, 0.05, 0.9]])
```

Tenemos a la matriz "P"  la cual modela el porcentaje de residentes que cambia de producto de un mes a otro, por ejemplo
P con fila i, columna j, representa la cantidad de gente que pasa de comprar el producto j a comprar el producto i de un mes al otro.
Siguiendo esta logica, podemos notar que podemos tratar a P como una transformacion lineal de una matriz X, que modele la cantidad de residentes.

Nos dan el modelo de matriz X(n), que modela la cantidad de compradores de cada producto en funcion del n-ésimo mes. Siguiendo el modelo entregado, nos dicen que X(0) (osea la cantidad de compradores al inicio) se modela de la siguiente forma:

```python
X0 = sp.Matrix([[10000],
               [10000],
               [10000]])
```

A)  <br />  <br />
    PX(0) : Modela la cantidad de residentes que compran cada producto despues de un mes.
    P(PX(0)): Es la misma translacion aplicada dos veces, asi obtendremos la matirz X(2), osea la cantidad de gente que compra cada producto luego de dos meses
    P(P(PX(0))) : De igual forma que las anteriores, es la translacion aplicada por tercera vez, en otras palabras, X(3), la cantidad de gente que compra cada 
    producto despues de 3 meses <br /> <br />
B) <br />  <br /> 
    Siguiendo la logica vista en el item A, podemos definir la siguiente funcion para calcular la cantidad de compradores en un mes N : <br />

```python
# Funcion que recibe las matrices P (t.linal), X0 (compradores iniciales), y un int N, que representa la cantidad de meses a modelar.
def calcular_compradores (P: sp.Matrix, X0: sp.Matrix, n: int) ->sp.Matrix:
    return (P**n) * X0
```

Y ahora con esta funcion definida, solo debemos iterarla con los valores que nos solicitan:

```python
meses = [5, 10 , 15, 20, 25, 30, 35, 40, 45, 50]
valores_de_Xn = [calcular_compradores(P, X0, n) for n in meses]
```

C) <br />  <br />
Con la lista de matrices listas, somos capaces de usar la libreria matplotlib.pyplot para graficar este modelamiento:
<br />  <br />

```python
#Creamos el grafico 
plt.figure(figsize=(10, 5))
#Creamos cada linea del grafio extrayendo la informacion de las matrices calculadas anteriormente
for i in range (3):
    compradores = [float(valores[i]) for valores in valores_de_Xn]
    plt.plot(meses, compradores, label = f"Producto {i+1}")
    
#Configuramos los labels y modelo del grafico
plt.xlabel("Meses")
plt.ylabel("Cantidad de compradores")
plt.title("Evolución de compradores por mes")
plt.grid(True)
plt.legend()
plt.xticks(meses)
plt.yticks(np.arange(2000,18000, 1000))
plt.show()
```

D) <br />  <br />Observamos que el producto 3 se va en alza desde el quinto al décimo mes, mientras que el producto 2 y 3 ven una caida en compradores. Todos los productos
ven un estancamiento a partir del quinceavo mes al rededor de determinados valores.

E) <br />  <br />

Si consideramos X(0) = [[0], [30000], [0]], tan solo debemos aplicar las funciones ya definidas y seguir los mismos pasos para esta nueva matriz:

```python
X0_1 = sp.Matrix([[0],
                  [30000],
                  [0]])
```

Aplicamos la funcion para calcular X(n):

```python
valores_de_Xn_1 = [calcular_compradores(P, X0_1, n) for n in meses]
```

Ahora que tenemos los valores de Xn_1 para cada mes, creamos el grafico:

```python
#Creamos el grafico 
plt.figure(figsize=(10, 5))
#Creamos cada linea del grafio extrayendo la informacion de las matrices calculadas anteriormente
for i in range (3):
    compradores = [float(valores[i]) for valores in valores_de_Xn_1]
    plt.plot(meses, compradores, label = f"Producto {i+1}")
    
#Configuramos los labels y modelo del grafico
plt.xlabel("Meses")
plt.ylabel("Cantidad de compradores")
plt.title("Evolución de compradores por mes")
plt.grid(True)
plt.legend()
plt.xticks(meses)
plt.yticks(np.arange(2000,18000, 1000))
plt.show()
```

Vemos que a pesar de haber empezado con una distinta distribucion de compradores, con el pasar de los meses el comportamiento es similar para ambos casos, e incluso se estabilizan en los mismos valores. Una alza en el producto 3 hasta los 16.000, donde se estabiliza rondando el numero, y una baja para el producto 2 y 3, estabilizandose en 9000 y 5000 respectivamente.
