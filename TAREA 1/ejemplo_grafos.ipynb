```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.Graph()

# Añadir nodos
G.add_nodes_from([1, 2, 3, 4, 5])

# Añadir aristas sin etiquetas automáticas
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (1, 5)])

# Posiciones de los vértices para que se parezca al grafo original
pos = {1: (0, 1), 2: (1, 1), 3: (2, 2), 4: (3, 1), 5: (4, 4)}

# Crear manualmente etiquetas para las aristas
labels = {(1, 2): 'a', (2, 3): 'b', (3, 4): 'c', (4, 5): 'd', (1, 5): 'e'}

# Dibujar el grafo
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10)

# Dibujar etiquetas manuales en las aristas
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Mostrar el grafo
plt.show()
```
