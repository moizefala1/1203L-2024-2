Un **grafo dirigido** es una colección de $n$ puntos denominados vértices, denotados
por $V_1$, $V_2$, $\ldots$, $V_n$, junto con un número finito de aristas que unen distintos pares de vértices.  A diferencia de un grafo normal, en un grafo dirigido las aristas tienen sentido y para denotarlas ocupamos flechas.


La ruta de un vértice hacia otro se denomina trayectoria o cadena. La trayectoria $V_3\to V_2\to V_5$ que va de $V_3$ a $V_5$  se llama 2-cadena porque pasa por dos aristas. En general una trayectoria que atraviesa por $n$ aristas (y por lo tanto pasa por $n+1$ vértices) se
llama $n$-cadena. Un ciclo es una trayectoria cuyo vértice inicial y final son el mismo. Por ejemplo, en el grafo de la figura, tenemos el ciclo  $V_3\to V_4\to V_1\to V_3$ cuyo largo es 3. El ciclo anterior cumple que el ciclo se recorre en el sentido de las aristas. Si este no fuera el caso decimos que el ciclo es no dirigido. Por ejemplo $V_3\to V_2\to V_5\to V_4\to V_3$ es un ciclo no dirigido de largo 4.

```python
from sympy import *
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos (vértices)
nodos = ["v1", "v2", "v3", "v4", "v5"]
G.add_nodes_from(nodos)

# Agregar aristas con direcciones según el gráfico
aristas = [
    ("v2", "v1"),
    ("v3", "v1"),
    ("v3", "v2"),
    ("v3", "v4"),
    ("v4", "v5"),
    ("v1", "v4"),
    ("v1", "v5")
]

# Agregar las aristas
G.add_edges_from(aristas)

# Posición manual de los nodos para que se asemeje al grafo original
pos = {
    "v1": (0, 2),
    "v2": (-1, 1),
    "v3": (0, 0),
    "v4": (1, 1),
    "v5": (2, 1)
}

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_size=700, node_color="white", edgecolors="black")

# Dibujar las aristas
nx.draw_networkx_edges(G, pos, edgelist=aristas, arrowstyle="->", arrowsize=20)

# Dibujar etiquetas de nodos
nx.draw_networkx_labels(G, pos, font_size=12, font_color="black")

# Mostrar el grafo
plt.title("Grafo Dirigido")
plt.axis("off")
plt.show()
```

Para este problema consideraremos que los grafos dirigidos satisfacen las siguientes dos condiciones:

1. Ningún vértice está conectado consigo mismo.
2. A lo más una arista lleva de un vértice a otro.


La matriz que representa una grafo dirigido que satisface estas dos condiciones se denomina **matriz de incidencia nodo-arista**. Se define como:

$$a_{ij} = \begin{cases}
1, & \text{si la arista $j$ llega al nodo $i$} \\
-1, & \text{si la arista $j$ sale del nodo $i$} \\
0, & \text{de otra manera}
\end{cases}$$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos (vértices)
nodos = ["v1", "v2", "v3", "v4", "v5"]
G.add_nodes_from(nodos)

# Agregar aristas (con direcciones según el gráfico)
aristas = [
    ("v1", "v2", "A1"),
    ("v2", "v4", "A2"),
    ("v4", "v1", "A3"),
    ("v1", "v3", "A4"),
    ("v3", "v5", "A5"),
    ("v5", "v1", "A6"),
    ("v3", "v4", "A7"),  # Arista adicional de v3 a v4
    ("v5", "v4", "A8"),
]

# Agregar las aristas con etiquetas
for (u, v, label) in aristas:
    G.add_edge(u, v, label=label)

# Posición manual de los nodos para que se asemeje al grafo original
pos = {
    "v1": (0, 2),
    "v2": (-1, 1),
    "v3": (1, 1),
    "v4": (-0.5, 0),
    "v5": (0.5, 0)
}

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_size=700, node_color="lightblue")

# Dibujar las aristas
nx.draw_networkx_edges(G, pos, edgelist=[(u, v) for (u, v, _) in aristas], arrowstyle="->", arrowsize=20)

# Dibujar etiquetas de nodos
nx.draw_networkx_labels(G, pos, font_size=12, font_color="black")

# Dibujar etiquetas de aristas
edge_labels = {(u, v): label for (u, v, label) in aristas}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, label_pos=0.3)

# Mostrar el grafo
plt.title("Grafo Dirigido")
plt.axis("off")
plt.show()
```

Un ciclo cerrado en este grafo puede ser, por ejemplo:

- $v1 \to v2 \to v4 \to v1$  (aristas A1, A2, A3)

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos (vértices)
nodos = ["v1", "v2", "v3", "v4", "v5"]
G.add_nodes_from(nodos)

# Agregar aristas (con direcciones según el gráfico)
aristas = [
    ("v1", "v2", "A1"),
    ("v2", "v4", "A2"),
    ("v4", "v1", "A3"),
    ("v1", "v3", "A4"),
    ("v3", "v5", "A5"),
    ("v5", "v1", "A6"),
    ("v3", "v4", "A7"),  # Arista adicional de v3 a v4
    ("v5", "v4", "A8")
]

# Agregar las aristas con etiquetas
for (u, v, label) in aristas:
    G.add_edge(u, v, label=label)

# Definir el ciclo y colorear las aristas del ciclo en rojo
ciclo1 = [("v1", "v2"), ("v2", "v4"), ("v4", "v1")]
ciclo2 = [("v1", "v3"), ("v3", "v5"), ("v5", "v1")]

# Posición manual de los nodos para que se asemeje al grafo original
pos = {
    "v1": (0, 2),
    "v2": (-1, 1),
    "v3": (1, 1),
    "v4": (-0.5, 0),
    "v5": (0.5, 0)
}

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_size=700, node_color="lightblue")

# Dibujar todas las aristas en gris claro inicialmente
nx.draw_networkx_edges(G, pos, edgelist=[(u, v) for (u, v, _) in aristas], edge_color="lightgray", arrowstyle="->", arrowsize=20)

# Dibujar las aristas de los ciclos en colores diferentes
nx.draw_networkx_edges(G, pos, edgelist=ciclo1, edge_color="red", width=2.5, arrowstyle="->", arrowsize=20, label="Ciclo 1")
nx.draw_networkx_edges(G, pos, edgelist=ciclo2, edge_color="green", width=2.5, arrowstyle="->", arrowsize=20, label="Ciclo 2")

# Dibujar etiquetas de nodos
nx.draw_networkx_labels(G, pos, font_size=12, font_color="black")

# Dibujar etiquetas de aristas
edge_labels = {(u, v): label for (u, v, label) in aristas}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, label_pos=0.3)

# Mostrar el grafo
plt.title("Grafo Dirigido con Ciclos Cerrados")
plt.axis("off")
plt.show()
```

c) Considere un subconjunto de aristas que no contengan ciclos cerrados. Pruebe la dependencia o independencia de las columnas correspondientes de A.

**Solución:**

Por ejemplo, podríamos elegir las aristas: $A1$, $A4$, $A7$, y $A8$. Este conjunto no incluye aristas suficientes para formar un ciclo, ya que carece de los enlaces de retorno necesarios.

Vamos a construir una submatriz de  A  tomando las columnas correspondientes a las aristas seleccionadas. En este caso, las aristas son $A1$, $A4$, $A7$, y $A8$, que corresponden a las columnas 1, 4, 7, y 8 de la matriz  A .

```python
# Seleccionar las columnas correspondientes a A1, A2, A7 y A8
submatriz = A[:, [0, 3, 6, 7]]


# Calcular el rango de la submatriz
rango = submatriz.rank()

# Verificar si las columnas son linealmente independientes
if rango == submatriz.shape[1]:
    print("Las columnas son linealmente independientes.")
else:
    print("Las columnas son linealmente dependientes.")
```

Si hacemos lo mismo, pero con las aristas de un ciclo obtenemos lo siguiente:

```python
# Seleccionar las columnas correspondientes a A1, A2, A3
submatriz = A[:, [0, 1, 2]]


# Calcular el rango de la submatriz
rango = submatriz.rank()

# Verificar si las columnas son linealmente independientes
if rango == submatriz.shape[1]:
    print("Las columnas son linealmente independientes.")
else:
    print("Las columnas son linealmente dependientes.")
```

Interpretación del Resultado

-	Linealmente Independientes: Si el rango de la submatriz es igual al número de columnas (4 en este caso), significa que las columnas seleccionadas son linealmente independientes. Esto implica que las aristas seleccionadas forman un conjunto de caminos independientes sin ciclos.

-	Linealmente Dependientes: Si el rango es menor que el número de columnas, esto indica dependencia lineal, lo cual podría suceder si accidentalmente elegimos aristas que aún forman un ciclo o son redundantes en términos de dirección.

Esta verificación asegura que el subconjunto de aristas seleccionado no contiene ciclos y que las columnas correspondientes de la matriz de incidencia son independientes.

elemento del vector corresponde al coeficiente de una arista. Por ejemplo, un ciclo en grafo dirigido anterior es: inicio en el nodo $V_3$, luego arista 5, después por la arista 8 y por el
opuesto de la arista 7. Esto se puede expresar como $A_5+A_8-A_7$, que se

-e1) Verifique que este vector está en el espacio nulo de A, la matriz de incidencia nodo-arista.

-e2) Forme el vector correspondiente al ciclo que va del nodo [1] al nodo [2] al nodo
[4] al nodo [3] y de regreso al nodo [1]. Verifique que este vector se encuentra en el
espacio nulo de A.


**Solución:**

Para resolver estos problemas, verificaremos si los vectores de ciclo están en el espacio nulo de la matriz de incidencia  A . Si un vector está en el espacio nulo de  $A$ , entonces el producto  $A \times v = 0$ , donde  v  es el vector de ciclo.

Parte  e1)

El primer vector de ciclo dado es  $[0, 0, 0, 0, 1, 0, -1, 1]^T$ , que representa el ciclo:

Inicio en  $v_3$ , a través de  $A_5$, luego a  $A_8$ y de regreso a través de  $-A_7$ .

Verificación:

Multiplicamos este vector por la matriz de incidencia  A  para verificar si está en el espacio nulo.

Parte  e2)

El segundo ciclo que debemos representar va de: $v_1$  a  $v_2$  por  $A_1$ , Luego de  $v_2$  a  $v_4$  por  $A_2$, Luego de  $v_4$  a  $v_3$  por  $-A_7$ y  de regreso a  $v_1$  por  $-A_4$.

Este ciclo se puede representar por el vector  $[1, 1, 0, -1, 0, 0, -1, 0]^T$.

Todo lo anterior se desarrolla con el siguiente codigo.


```python
import sympy as sp
# Parte e1: Definir el vector de ciclo [0, 0, 0, 0, 1, 0, -1, 1]
v_ciclo_1 = sp.Matrix([0, 0, 0, 0, 1, 0, -1, 1])

# Verificar si está en el espacio nulo multiplicando por A
resultado_1 = A * v_ciclo_1

# Parte e2: Definir el vector de ciclo [1, 1, 0, -1, 0, 0, -1, 0]
v_ciclo_2 = sp.Matrix([1, 1, 0, -1, 0, 0, -1, 0])

# Verificar si está en el espacio nulo multiplicando por A
resultado_2 = A * v_ciclo_2

# Mostrar los resultados
print("Resultado de A * v_ciclo_1:", resultado_1)
print("Resultado de A * v_ciclo_2:", resultado_2)
```
